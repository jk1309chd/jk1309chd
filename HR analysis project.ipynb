## Using IQR method for outlier treatment

```python
Q1= df['time_spend_company'].quantile(0.25)
Q3= df['time_spend_company'].quantile(0.75)
IQR= Q3-Q1
LD=Q1-1.5*IQR
UBD=Q3+1.5*IQR
upper_bound=np.where(df['time_spend_company']<=LD)[0]
Lower_bound =np.where(df['time_spend_company']>=UBD)[0]
df1=df.drop(index=upper_bound)
df1=df.drop(index=Lower_bound)
df1

```

## Training the Logistic regression model for Turnover prediction

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test=train_test_split(X1,y1,test_size=.20,random_state=123)
LR=LogisticRegression()
model1=LR.fit(X1_train,y1_train)
model1
```

```python
prediction=model1.predict(X1_test)
prediction
```

## Accuracy score & classification report of Logistic regression

```python
from sklearn.metrics import accuracy_score , confusion_matrix, classification_report, ConfusionMatrixDisplay
acc=accuracy_score(y1_test,prediction)
acc
```

```python
print(confusion_matrix(y1_test,prediction))  ##C00 = true neg, C10=false neg, C11= true +ve, C01 = false +ve

print(classification_report(y1_test,prediction))
```

## Confusion matrix for Logistic regression

```python
plt.figure(figsize=(15,8))
ConfusionMatrixDisplay.from_predictions(y1_test,prediction)
plt.show()
```

## Based on the above probability score range, categorize the employees into four zones 

```python
def category(x):
    if (x <20):
        x="Safe zone"
    elif (x >=20) & (x<=60):
        x="low risk zone"
    elif (x >60) & (x<=90):
        x="medium risk zone"
    else:
        x="high risk"    
    return x
```

```python
X1_test['prediction_category']=X1_test['prediction_category'].apply(category)
```

## ■	Medium Risk Zone (Orange) (60% < Score < 90%): Working hours are signficantly high which should be checked and redused appropratly and this group has low income and needed to be moved from low to medium group
